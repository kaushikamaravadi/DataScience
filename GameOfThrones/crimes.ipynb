```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import operator
import warnings
%matplotlib inline
sns.set_style('white')
from pandas import Series, DataFrame
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("report.csv")
```

```python
df
```

```python
unclean = df[df.agency_jurisdiction == 'United States']
df=df[df.agency_jurisdiction != 'United States']
```

```python
all_crimes=plt.figure(figsize=(10,22))
all_crimes = sns.barplot(df["violent_crimes"]+df["homicides"]+
                 df["rapes"]+df["assaults"]+df["robberies"],                 
                 y=df["agency_jurisdiction"],estimator=sum ,ci=0)
all_crimes.set_title('Total crimes in US during (1975-2015)')
all_crimes.set(xlabel='Total number of crimes', ylabel='City')
plt.show()
```

```python
plt.figure(figsize=(18,4))
sns.barplot(df["report_year"], df["crimes_percapita"], palette="BuGn_d",)
plt.ylabel("Rate of crimes")
plt.title('Crimes rates in New York City in 1975-2015')
plt.xlabel("Report Year")
```

```python
plt.figure(figsize=(18,4))
sns.barplot(df["report_year"], df["violent_crimes"], palette="RdBu_r")
plt.ylabel("Total crimes in USA")
plt.title('Crimes in USA in 1975-2015')
plt.xlabel("Report Year")
plt.show()
```
